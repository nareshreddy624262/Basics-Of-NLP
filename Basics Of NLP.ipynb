```python
voc = []
stem_word = []
for i in sent_tokenize(text):
    for j in word_tokenize(i):
        lower = j.lower()
        if lower not in stopwords.words("english") and len(lower) >= 2:
            voc.append(lower)
stem = PorterStemmer()
for k in voc:
    stem_word.append(stem.stem(k))
```
